# 1. Scatter plot game
*   Randomly generate points on a graph and the player has to input the (x,y) coordinates
*   For added difficulty, make the graph larger



```python
import matplotlib.pyplot as plt
import numpy as np
import random
def easy():
  is_playing = True
  while is_playing:
    x, y = random.randint(-10,10), random.randint(-10,10)
    fig, ax = plt.subplots()
    plt.axis([-10,10,-10,10])
    plt.plot([-10,10],[0,0],'g')
    plt.plot([0,0],[-10,10],'g')
    plt.title('Scatter plot game')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.scatter([x], [y], color='purple', s=75)
    plt.xticks(np.arange(-10,10,2))
    plt.yticks(np.arange(-10,10,2))
    plt.grid(linestyle='dotted', color='yellowgreen')
    plt.show()
    guess = input("enter x and y (separated with a comma):")
    my_guess = guess.split(',')
    if int(my_guess[0]) == x and int(my_guess[1]) == y:
      print('Correct!')
    else:
      print('Wrong!')
    play_again = input('play again? (y/n) ')
    if play_again == 'n':
      print('Thanks for playing')
      is_playing = False
      break
def difficult():
  is_playing = True
  while is_playing:
    x, y = random.randint(-20,20), random.randint(-20,20)
    fig, ax = plt.subplots()
    plt.axis([-20,20,-20,20])
    plt.plot([-20,20],[0,0],'g')
    plt.plot([0,0],[-20,20],'g')
    plt.title('Scatter plot game')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.scatter([x], [y], color='purple', s=75)
    plt.xticks(np.arange(-20,20,2))
    plt.yticks(np.arange(-20,20,2))
    plt.grid(linestyle='dotted', color='yellowgreen')
    plt.show()
    guess = input("enter x and y (separated with a comma):")
    my_guess = guess.split(',')
    if int(my_guess[0]) == x and int(my_guess[1]) == y:
      print('Correct!')
    else:
      print('Wrong!')
    play_again = input('play again? (y/n) ')
    if play_again == 'n':
      print('Thanks for playing')
      is_playing = False
      break
choice = input('easy or difficult? (e/d) ')
if choice == 'e':
  easy()
elif choice == 'd':
  difficult()
```

# 2. Algebra practice game
*   Generate one-step and two-step problems with random integer values and the player has to input the answer
*   Use positive and negative values. For added difficulty, make the numbers larger

```python
import random
def easy():
  is_solving = True
  while is_solving:
    num1 = random.randint(-5,5)
    num2 = random.randint(-5,5)
    x = random.randint(-5,5)
    y1 = num1 * x + num2
    y2 = x * num1 - num2
    print('****Algebra practice game****')
    print('Problem 1:')
    print(f'{num1} * x + {num2} = {y1}')
    guess_int1 = int(input('Enter an integer number: '))
    if guess_int1 == x:
      print('Correct!')
    else:
      print(f'Wrong! The correct answer is {x}')
    print('Problem 2:')
    print(f'x * {num1} - {num2} = {y2}')
    guess_int2 = int(input('Enter an integer number: '))
    if guess_int2 == x:
      print('Correct!')
    else:
      print(f'Wrong! The correct answer is {x}')
    solve_math = input('solve math (yes/no)? ')
    if solve_math == 'n':
      print('Thanks!')
      is_solving = False
      break
def difficult():
  is_solving = True
  while is_solving:
    num1 = random.randint(-10,10)
    num2 = random.randint(-10,10)
    x = random.randint(-10,10)
    y1 = num1 * x + num2
    y2 = x * num1 - num2
    print('****Algebra practice game****')
    print('Problem 1:')
    print(f'{num1} * x + {num2} = {y1}')
    guess_int1 = int(input('Enter an integer number: '))
    if guess_int1 == x:
      print('Correct!')
    else:
      print(f'Wrong! The correct answer is {x}')
    print('Problem 2:')
    print(f'x * {num1} - {num2} = {y2}')
    guess_int2 = int(input('Enter an integer number: '))
    if guess_int2 == x:
      print('Correct!')
    else:
      print(f'Wrong! The correct answer is {x}')
    solve_math = input('solve math (yes/no)? ')
    if solve_math == 'n':
      print('Thanks!')
      is_solving = False
      break
choice = input('easy or difficult? (e/d) ')
if choice == 'e':
  easy()
elif choice == 'd':
  difficult()
```

# 3. Projectile game
*   Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall
*   For added difficulty, make a second level where players enter a, b, and c without sliders

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
import random
from IPython.display import display
# projectile game
location = random.randint(2,18)
height = random.randint(2,18)
def easy():
  def projectile_game(a,b,c):
    xmin = 0
    xmax = 20
    points = 10*(xmax-xmin)
    x = np.linspace(xmin,xmax,points)
    y = a*x**2 + b*x + c
    shot_wall = a*location**2 + b*location + c
    x_limit = xmin,xmax
    y_limit = xmin,xmax
    plt.figure(figsize=(8,6))
    plt.xlim(x_limit)
    plt.ylim(y_limit)
    plt.axhline(0,color='black',linewidth=5)
    plt.bar(location,height,label='wall',width=0.75)
    plt.scatter(location,height + 1,label='height',color='r',s=100)
    plt.plot(x,y,label='path',color='magenta',linewidth=3)
    plt.title(f"Projectile game: {'Success' if shot_wall > height else 'Failed'}")
    plt.legend()
    plt.grid(linestyle='dotted', color='g')
    plt.show()
  iplot = interactive(projectile_game,a=(-5.0, 0.0, 0.5), b=(0.0, 20.0, 0.25), c=(0.0, 20.0, 0.25))
  display(iplot)
def difficult(a,b,c):
  xmin = 0
  xmax = 20
  points = 10*(xmax-xmin)
  x = np.linspace(xmin,xmax,points)
  y = a*x**2 + b*x + c
  shot_wall = a*location**2 + b*location + c
  x_limit = xmin,xmax
  y_limit = xmin,xmax
  plt.figure(figsize=(8,6))
  plt.xlim(x_limit)
  plt.ylim(y_limit)
  plt.axhline(0,color='black',linewidth=5)
  plt.bar(location,height,label='wall',width=0.75)
  plt.scatter(location,height + 1,label='height',color='r',s=100)
  plt.plot(x,y,label='path',color='magenta',linewidth=3)
  plt.title(f"Projectile game: {'Success' if shot_wall > height else 'Failed'}")
  plt.legend()
  plt.grid(linestyle='dotted', color='g')
  plt.show()
choice = input('easy or difficult? (e/d) ')
if choice == 'e':
  easy()
elif choice == 'd':
  a = int(input('num1: '))
  b = int(input('num2: '))
  c = int(input('num3: '))
  difficult(a,b,c)
```
